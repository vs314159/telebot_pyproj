```python
def extract_button_text(file, keyword):
    with open(file, mode='r', encoding='utf-8') as f:
        result = tuple(map(lambda x: x.split(':')[1].strip('\n "'), filter(lambda x: keyword in x, f)))
    return result
```

```python
def get_keys(graph):
    res = []
    for key in graph.keys():
        if key not in res:
            res.append(key)
        res += graph[key]
    return res
```
